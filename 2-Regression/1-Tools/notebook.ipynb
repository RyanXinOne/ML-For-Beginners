# Linear Regression on diabetes dataset

Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
# model_selection is used for splitting data into training and test sets
from sklearn import datasets, linear_model, model_selection
```

Load the diabetes dataset

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

**Feature Selection**: Select a portion of this dataset to plot by arranging it into a new array using numpy's `newaxis` function

```python
X = X[:, np.newaxis, 2]     # Equivalent to X = X[:, 2:3]
X.shape
```

**Hold-Out Validation**: Split both the data (`X`) and the target (`y`) into test and training sets

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
print(X_train.shape, y_test.shape)
```

Train the linear regression model

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

Create a prediction using test data

```python
y_pred = model.predict(X_test)
y_pred.shape
```

Create scatterplot of test data and draw the prediction line

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```

## Linear vs. NonlinearCategorical Variables

If a researcher has a good reason to suspect that the effect of an independent variable is consistent for all values of that variable, then it's perfectly fine to treat that variable linearly. If, on the other hand, the researcher has a good reason to think that the effect of that variable will change depending on its value, it makes more sense to create a set of dummies.
